## モデル内の要素を、一切合切プロットする

* モデル内の要素を、ひとつひとつ指定して記録するのではなく、任意の条件に該当する要素をリストアップして、そのすべてを記録し、グラフにプロットしてみます。
* それぞれの `Entity` には、必ず備わっているデフォルトの属性があります。`Variable` であれば現在量を示す `Value` 属性、`Process` であれば、現在の反応速度を示す `Activity` 属性などがあります。今回は、これらを対象に、リストアップし、`Logger` をつくり、プロットしてみます。


### 対象の `LoggerStub` をつくり、辞書に格納する

* `System:/:cell` 内のすべての `Varieble` の `Value` 属性、すべての `Process` の `Activity` 属性について、`LoggerStub` をつくり、`FullPN` をキーとする **辞書 `Logger_dict`** に納める

```python
target_SystemPath = '/cell'
Target_Properties = { 'Variable':'Value', 'Process':'Activity' }
Logger_dict = {}

for E_type, p in Target_Properties.items():
    for E in getEntityList( E_type, target_SystemPath ):
        FullPN = ':'.join( ( E_type, target_SystemPath, E, p ) )
        Logger_dict[ FullPN ] = createLoggerStub( FullPN )
```

###  `Logger_dict` 内のすべての `LoggerStub` に対して、 `create()` を実行し、記録を開始する

```python
for L in Logger_dict.values():
    L.create()
```

### データを辞書に格納する

* データを、`FullPN` をキーとする **辞書 `Data_dict`** に納める
* このモデルでは、大部分の  `Value`  は頻度を表すため 0 ≦  `Value`  ≦ 1 の範囲内にある。一方、 `Process`  はその限りではない。
* `Value` と `Process` を分別できるよう、それぞれの `FullPN` を格納したリスト `Variable_FullPN_list` 、`Process_FullPN_list` を作成する。

```python
Data_dict = {}
for FullPN, Logger in Logger_dict.items():
    Data_dict[ FullPN ] = np.array( Logger.getData( 0, 100, 0.1 ) )[ :, :2 ]

Variable_FullPN_list = [ FullPN for FullPN in Data_dict.keys() if FullPN.split(':')[ 0 ] == 'Variable' ]
Process_FullPN_list = [ FullPN for FullPN in Data_dict.keys() if FullPN.split(':')[ 0 ] == 'Process' ]
```

* `Variable` のデータを１つの `figure` にプロットする。


```python
for FullPN, d in Data_dict.items():
    if FullPN in Variable_FullPN_list:
        bokeh.plotting.line( d[ :, 0 ], d[ :, 1 ], line_width = 1.0 )

bokeh.plotting.show()
```

```python
from bokeh.plotting import *
from bokeh.embed import autoload_server
from bokeh.session import Session

colors = ("#A6CEE3", "#1F78B4", "#B2DF8A")
N = 36
r_base = 8
theta = np.linspace(0, 2*np.pi, N)
r_x = np.linspace(0, 6*np.pi, N-1)
rmin = r_base - np.cos(r_x) - 1
rmax = r_base + np.sin(r_x) + 1

output_server("wedge animate")

cx = cy = np.ones_like(rmin)
annular_wedge(cx, cy, 
        rmin, rmax, theta[:-1], theta[1:],
        inner_radius_units="data",
        outer_radius_units="data",
        color = colors[0], 
        #line_color="black", tools="pan,box_zoom,resize,reset"
        )
#show()

import time
from bokeh.objects import Glyph
renderer = [r for r in curplot().renderers if isinstance(r, Glyph)][0]
ds = renderer.data_source
while True:
    for i in np.linspace(-2*np.pi, 2*np.pi, 50):
        rmin = ds.data["inner_radius"]
        rmin = np.roll(rmin, 1)
        ds.data["inner_radius"] = rmin
        rmax = ds.data["outer_radius"]
        rmax = np.roll(rmax, -1)
        ds.data["outer_radius"] = rmax
        ds._dirty = True
        cursession ().store_objects(ds)
        time.sleep(.25)
```

```python
# 毎回 figure() を実行すると、その都度新たな figure がつくられて別々のプロットになる。

for FullPN, d in Data_dict.items():
    if FullPN in Process_FullPN_list:
        plt.figure()
        plt.plot( d[ :, 0 ], d[ :, 1 ], label = FullPN.split(':')[ 2 ] )
        plt.legend( loc = 'upper right' )
```
