# グループワーク課題１

コンピュータ上で、簡単なモデルを使った**生存競争**のシミュレーションを行います。  
このシミュレーションには、各グループ1000個体、合計4000個体が参加します。

**「シミュレーションに参加する個体のプログラムを作る」のが、各グループの課題です。**

- 合計1000個体であれば、何種類の個体を混ぜても構いません。

## 生存競争モデル

- 個体が生存競争するフィールドには、**〔豊かな場所〕**があります。〔豊かな場所〕には集団の半分（2000個体）しか入れません。
- そこで、〔豊かな場所〕にいる2000個体と、そうでない2000個体の間で〔豊かな場所〕をめぐる**闘争**が起こります。
- 闘争は１対１で起こります。つまり、2000組の１対１の闘争が起こります。
- 生存競争は、以下の手順で進めていきます。


1. 各グループは１つまたはそれ以上のタイプの個体を用意し、その合計が1000個体になるようにします。複数のタイプがある場合、それぞれのタイプの個体数は自由に割り当てて構いません。４グループあるので、合計4000個体になります。これを**初期集団**とします。

2. 4000個体からランダムに選んだ2000個体を〔豊かな場所〕に配置します。

3. 〔豊かな場所〕にいる2000個体と、そうでない2000個体をランダムに組み合わせて2000組の闘争の組み合わせを決めます。どのグループがつくったかに関わらず、ランダムに組み合わせるので、同士討ちも起こります。

4. 闘争の際に、各個体は、表現型 $F$ または $M$ のどちらかを出します。組み合わせによって、各個体はポイントを得ます。得られるポイントは下表の通りです。

|表現型|相手が$F$|相手が$M$|
|:-:|:-:|:-:|
|$F$|-1|1|
|$M$|0|0.5|

  - $F$ と $M$ が戦うと、つねに $F$ が勝ちます。ただし、$M$ は抵抗せずに逃げるため、$F$ は１ポイントを得て、$M$ は±０ポイントです。
  - $F$ どうしは、激しく戦って傷つき、痛み分けで双方 -１ ポイントです。
  - $M$ どうしは、平和的に資源を分けあって、双方が 0.5 ポイントを得ます。

<ol start="5"><li>
個体のタイプ毎に、闘争で得たポイントの合計（**タイプの総ポイント**）を算出します。
</li><li>
タイプの総ポイントに比例して、次の世代での個体数を決めます。集団のサイズ（個体数）は、毎世代4000個体です。あるタイプの総ポイントが $x$ で、すべてのタイプの総ポイントの合計が $\Sigma$ のとき、そのタイプの次世代の個体数は、$\frac{4000 x}{\Sigma}$ となります（四捨五入して整数にします）。
</li><li>
次世代の集団の構成が決まったら、ステップ2〜6を繰り返します。これを繰り返して**世代発展**をシミュレートします。
</li><li>
集団全体がある１つのタイプに独占されたら、シミュレーションを終了します。  
また、ある周期で同じ状況に戻ってくるような**周期定常状態**になった場合も、シミュレーションを打ち切ります。
</li></ol>

## 個体のタイプのプログラム

個体のタイプは、Python の関数として実装します。

    def type( position, memory = "☆" ):
        （コード）
        return "○"

- **引数**
  - `position` 〔豊かな場所〕にいる場合 1, いない場合は 0
  - `memory`　この個体の〔親〕が、前の世代で発現した表現型。〔親〕は、前の世代の同じタイプの個体からランダムに選んだ１個体とします。最初の世代の値は "`☆`" です。
- **戻り値** "`F`" または "`M`"（その個体が発現する表現型）

他の情報（自分と同じタイプの個体の数など）は一切利用できない（個体が知ることはできない）ものとします。  
引数で得られる情報を使って使わなくても自由です。$F$ か $M$ のいずれかを決め打ちで出しつづけるだけのタイプでも構いません。



```python
# 提出するプログラム　（複数ある場合は列挙してください）

# 以下は例です。

# 初期個体数 400
def type_1( position, memory = "F" ):
    if memory == "F":
        return "M"
    else:
        return "F"

# 初期個体数 600
def type_2( position, memory = "F" ):
    if position == 1:
        return "F"
    else:
        return "M"
```
