## E-Cell によるシミュレーションの実行

E-Cell を用いて、モデルファイルの読み込み、記録するデータの設定、シミュレーションの実行、記録したデータの可視化を行います。

### モデルファイルの読み込み

モデルファイルを読み込みます。E-Cell3 のモデルファイルは、EML（E-Cell Model Launguage）形式で書かれています。

ここで読み込むモデルファイルは、データベース [BioModels](http://www.ebi.ac.uk/biomodels-main/ "BioModels") に収録されているモデルから生成されたものです。

loadModel( _モデルファイル名_ ) は、モデルファイルを読み込む関数です（Extension によって IPython に追加された機能のひとつです）。

```python
loadModel("BIOMD0000000005.eml")
```

モデルの詳細は、BioModels で閲覧できます：[BIOMD0000000005](http://www.ebi.ac.uk/biomodels-main/BIOMD0000000005 "BIOMD0000000005 on BioModels")

ヴァージニア工科大学の Tyson による細胞周期（cdc2 とサイクリンの相互作用）のモデルです。

参考文献

* Tyson JJ. Modeling the cell division cycle: cdc2 and cyclin interactions. _Proc Natl Acad Sci U S A._ 1991 Aug 15;__88__(16):7328-32. [PubMed PMID: 1831270](http://www.ncbi.nlm.nih.gov/pubmed/1831270 "PubMed")

### 記録するデータの設定

モデルから記録するデータを選び、**Logger**を作成します。

ここでは、YT（サイクリンの総量）、M（リン酸化サイクリン-cdc2複合体）、CT（cdc2の総量）の３要素のLoggerを作成します。

createLoggerStub() はLoggerにアクセスする足がかり（Stub）を作成する関数で、この時点で、Loggerそのものはつくられません（記録は始まりません）。

```python
YT = createLoggerStub( 'Variable:/cell:YT:Value' )
M = createLoggerStub( 'Variable:/cell:M:Value')
CT = createLoggerStub('Variable:/cell:CT:Value')
```

_LoggerStubオブジェクト_.create() を実行することでLoggerがつくられ、記録が始まります。

二段階に分かれているのは一見面倒ですが、大規模なシミュレーションでは、膨大なデータを保存する必要があり、必要とされる時点でLoggerによる記録を開始するのが、効率的です（速度面でも高速）。そこで、必要となるLoggerを準備する段階と、記録を始める段階を分けて記述するように設計されています。

```python
YT.create()
M.create()
CT.create()
```

### シミュレーションの実行

run( t ) を実行すると、時間 _t_ だけシミュレーションを実行します。_t_ の単位はモデルによります。

```python
run(100)
```

### データの可視化

シミュレーションで得たデータを可視化してみます。

#### シミュレーション結果を numpy の array 形式に変換する

_LoggerStubオブジェクト_.getData( _開始_, _終了_, _間隔_ ) で、Loggerの記録内容から、_開始_時間から、_終了_時間までのデータを、_間隔_の時間間隔で取りだすことができます。

numpy は Python でデータ分析を行う際によく用いられるモジュールで、E-Cell3 でも使われています。ここでは、numpy に **np** という別名を与えています。別名を与えたことで、例えば np.array() という関数を呼ぶのに、np.array() と書けばよくなります。頻繁に使うモジュールに単純な別名を与えることによって、コードを短く、すっきりとさせることができます。

np.array( _data_ ) は、引数に与えられたデータ _data_ を numpy のデータ形式（array＝配列）に変換する関数です。

```python
import numpy as np

YT_Data = np.array( YT.getData(0,100,0.1) )
M_Data = np.array( M.getData(0,100,0.1) )
CT_Data = np.array( CT.getData(0,100,0.1) )
```

#### Matplotlib を用いてグラフを描く

ここでは、グラフを書いてみます。ここでは、Matplotlibという高機能な可視化ライブラリから、さまざまなグラフ描画を行えるモジュール **matplotlib.pyplot** を用います。

マジックコマンド %matplotlib inline は、webページ上に結果の画像を表示させるコマンドです。

import matplotlib.pyplot as plt で、matplotlib.pyplot に **plt** という別名を与えます。


```python
%matplotlib inline
#from pylab import *
import matplotlib.pyplot as plt
```

さらに、MATLABという科学技術計算ソフトウェアに似た簡単なインターフェイスを実現してくれるモジュール **pylab** をインポートします。

```python
from pylab import *
```

簡単なグラフ描画をしてみます。

```python
figure()
plot( CT_Data[ :, 0 ], YT_Data[ :, 1 ] / CT_Data[ :, 1 ] )
xlabel( 'time' )
ylabel( 'YT/CT' )
title( 'YT/CT' )
```

グラフを並べたり、色を付けたり、点線にしたりできます。

```python
subplot(1,2,1)
plot( CT_Data[ :, 0 ], YT_Data[ :, 1 ] / CT_Data[ :, 1 ], 'r--' )
subplot(1,2,2)
plot( CT_Data[ :, 0 ], M_Data[ :, 1 ] / CT_Data[ :, 1 ], 'g*-')
```

pylab を使わずに、細部までグラフを設計してみます。

matplotlib では、figure という〔カンバス〕を作成して、その上に（ひとつまたは複数の）グラフ（axes）を描くという概念モデルを採っています。


```python
fig = plt.figure()

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot( CT_Data[ :, 0 ], YT_Data[ :, 1 ] / CT_Data[ :, 1 ], label="YT/CT")
ax.plot( CT_Data[ :, 0 ], M_Data[ :, 1 ] / CT_Data[ :, 1 ],  label="M/CT")

# 凡例
ax.legend( loc = 2 )

# 軸ラベル, グラフのタイトル
ax.set_xlabel('time')
ax.set_ylabel('concentration')
ax.set_title('Cyclin, cdc2 dynamics')
```

axes を複数作って、ひとつの figure 上に、複数の axes を重ねることもできます。

```python
fig = plt.figure()

# main figure
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes

axes1.plot( YT_Data[:,0], YT_Data[ :, 1 ] / CT_Data[ :, 1 ], 'r')
axes1.set_xlabel('time')
axes1.set_ylabel('YT/CT')
axes1.set_title('YT/CT')

# insert
axes2 = fig.add_axes([0.2, 0.2, 0.4, 0.3]) # inset axes

axes2.plot(CT_Data[ :, 0 ], M_Data[ :, 1 ] / CT_Data[ :, 1 ], 'g')
axes2.set_xlabel('time')
axes2.set_ylabel('Cyclin-cdc2/CT')
axes2.set_title('Cyclin-cdc2/CT')
```

YT と M の **相空間プロット** です。

```python
fig = plt.figure()

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot( YT_Data[ :, 1 ], M_Data[ :, 1 ], 'r', label="YT/M")
ax.legend( loc = 2 )
ax.set_xlabel('total cyclin')
ax.set_ylabel('cyclin-cdc2 complex')
ax.set_title('phase space plot')
```
