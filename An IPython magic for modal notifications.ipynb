```python
import os

def check_suffix( f, suffixes ):
    for suffix in suffixes:
        if f.endswith( '.{}'.format( suffix )):
            return True
    return False

def list_modelfile( suffixes = [ 'em', 'eml' ] ):
    pwd = os.getcwdu() 
    files_in_pwd = os.listdir( pwd )

    # Model Files
    modelfiles = [ f for f in files_in_pwd if check_suffix( f, suffixes ) ]
    
    # Directories
    dirs = [ f for f in files_in_pwd if ( os.path.isdir( pwd + '/' + f ) and ( not f.startswith('.') ) ) ]
    
    return ( dirs, modelfiles )

```
